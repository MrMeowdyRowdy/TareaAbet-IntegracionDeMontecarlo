```python
@njit
def f(x,r,a,b): #siendo x el rango a analizar, r el radio, a y b -> X y Y en coordenandas cartesianas respectivamente
    y = np.sqrt(r**2-x**2+2*a*x-a**2)+b #funcion para el semicirculo inferior
    return y;  # evaluamos fx

@njit
def fm(x,r,a,b): #siendo x el rango a analizar, r el radio, a y b -> X y Y en coordenandas cartesianas respectivamente
    y = -(np.sqrt(r**2-x**2+2*a*x-a**2)) + b #funcion para el semicirculo inferior
    return y;  # evaluamos fx
```

```python
def realizar_muestra(a, b, n=10):
    """
    a, b: intervalo
    n: tamaño de la muestra
    """
    return np.random.rand(n)*(b - a) + a


def evaluar_fx_muestra(m):
    """
    m: muestra (array)
    evaluamos la función f definida arriba
    """
    return f(m)
```
